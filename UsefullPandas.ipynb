```python
import pandas as pd
```

```python
def replace_empty_col_with(df, colname, value_to_replace=""):
    """Goes through the column 'colname' and replaces all empty fields (NaN) with value_to_replace"""
    df.loc[df[colname].isnull(), colname] = value_to_replace
    return df
```

```python

def split_column_entries_by_newline(df, colname_on_which_to_split):
    extra_cols = set(df.columns) - set(colname_on_which_to_split)
    
    df = replace_empty_col_with(df, colname='Mechanism', value_to_replace="") # make sure that the mechanism column does not contain NaN
    
    collector = list()
    for i, row in df.iterrows():
        for element in row[colname_on_which_to_split].split('\n'):
            processed_element = element.strip()
            if len(processed_element) > 1:
                new_row = {extra_col_name: row[extra_col_name] for extra_col_name in extra_cols} 
                new_row['Mechanism'] = processed_element
                collector.append(new_row)

    out = pd.DataFrame.from_records(collector)
    return out
```

```python
xx = pd.read_excel("../Vaude_TEST.xlsx")
final = split_column_entries_by_newline(xx, 'Mechanism')
```
